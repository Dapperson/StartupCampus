### 12.3. `__init__()` function

Semua class memiliki fungsi bernama `__init__()`, dimana fungsi ini akan diexecute ketika class diinisiasi.

Contoh:

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Zulfikar", 28)

print(p1.name)
print(p1.age)
```

# 13. Python Inheritance

Dengan inheritance, kita bisa membuat class baru yang inherits semua method dan properti dari class lain.

Contoh kita memiliki `parent class`:

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)


x = Person("Zulfikar Lazuardi", "Maulana")
x.printname()
```

```python
class Student(Person):
  pass
```

```python
y = Student("Jaka", "Tingkir")
y.printname()
```

```python
class Student(Person):
  def __init__(self, fname, lname, age):
    Person.__init__(self, fname, lname)
    self.age = age
  
  def printage(self):
    print(str(self.age))
```

```python
y = Student("Jaka", "Tingkir", 30)
y.printname()
```

```python
y.printage()
```

# 14. Python Iterators

Iterator adalah object yang berisi value yang bisa dihitung dan bisa diiterate ke semua value.

di python, iterator adalah object yang terdiri dari iterator protocol `__iter__()` dan `__next__()`

```python
mytuple = ("apple", "banana", "cherry")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

```python
class MyDummyClass:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    x = self.a
    self.a += 1
    return x

myclass = MyDummyClass()
myiter = iter(myclass)

print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```

```python
class MyDummyClass:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyDummyClass()
myiter = iter(myclass)

for x in myiter:
  print(x)
```
